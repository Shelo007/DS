Now, there is one surprising findings that I have found in this database that I'd like to share with you guys. Usually a video trends based on sharing! While doing that, one or the other person tends to like or dislike the video. Even if comments are disabled, people tend to react for the videos while sharing it and making it a viral one. But I have found some unique exceptional trending video statistics. There are few videos which trended in 2017 and 2018 for which the videos didn't get to receive even a single like or dislike. Also, their comments section were disabled by the channel. So, I have considered those channels which disabled the comments section as Non profit channels. Let's see!

```python
# Creating a dataframe for which channels have their comments disabled, and have got zero likes and zero dislikes 

Nonprofit =df1['comments_disabled']==1
Good =df1['likes']== 0 
Bad =df1['dislikes']== 0 
Surprising =df1[Nonprofit & Good & Bad]
```

```python
# Dropping one column named title which is not required for the analysis
df2=Surprising.drop(['title'],axis = 1)
df2
```

```python
#Now, we will first just look at the trending category of Non-profit motive channels
%matplotlib inline
plt.style.use('ggplot')

x = df2["Trending_Number"]
y = df2["views"]


plt.bar(x, y, color='green')
plt.xlabel("Trending Number")
plt.ylabel("Viewers Rate")
plt.title("Trending Videos for which comments are disabled")

plt.show()

#The plot shows a general decline in mortality of this cancer for this age group from 1999 to 2015. 
#The result here supports the hypothesis that mortality has decreased over time.
```

Here, we find that some videos were trending at 25 with more viewers and some videos trended at 2nd position in the entire country with less veiwers comparatively.
